```python
%matplotlib inline

import pandas as pd
import re
import requests

articles_list = []
for i in range(0,3):
    url = 'https://sitesearchapp.washingtonpost.com/sitesearch-api/v2/search.json?count=20&datefilter=displaydatetime:%5B*+TO+NOW%2FDAY%2B1DAY%5D&facets.fields=%7B!ex%3Dinclude%7Dcontenttype,%7B!ex%3Dinclude%7Dname&highlight.fields=headline,body&highlight.on=true&highlight.snippets=1&query=brexit&sort=&spellcheck=true&startat=' + str(i*20) + '&callback=angular.callbacks._1'
    page = requests.get(url)
    page_html = page.text
    articles_list= articles_list + re.findall('contenturl":"(.*?)"', page_html)
    
    
articles_list
```

```python
from newspaper import Article
def get_article_info(url):
    article = Article(url)
    article.download()
    article.parse()
    
    article_details = {'title'       : article.title,
                       'text'        : article.text,
                       'url'         : article.url,
                       'authors'     : article.authors,
                       'html'        : article.html,
                       'date'        : article.publish_date,
                       'description' : article.meta_description,
                       'publisher'   : article.meta_data['dc.publisher']}
    return article_details

article_details = []
for i in range(0,len(articles_list)):
    article_details.append(get_article_info(articles_list[i]))
```

```python
df_articles = pd.DataFrame(article_details)
```
